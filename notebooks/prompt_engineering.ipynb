# Prompts

## Few-shot classifier 

The prompt of the few-shot classifier contains 4 parts:

- General Instructions (*I want you to classify text for me. Your response must always be a python-like list of the 3 most possible class labels like: [0, 12, 15]  
See below all the possible labels and their description*)  
- All possible Intents and their label (unique id)  
- A few examples for each class (intent) with their label  
- Text to be classified with a prompt to provide the top 3 target labels as a python list    

A number of random examples per class (2 or 3) are passed to the few-shot classifier to provide more accurate class reference.  
I am using **jinja2** to define the backbone of my prompt and pass classes and examples dynamically. 

Before each request to openai the training queries & user_input are free of stopwords and special characters.  

Prompt template for few-shot classifier: [few-shot-template.jinja2](../data/few_shot_template.jinja2)

## Zero-Shot Classifier

The prompt of the zero-shot classifier contains 3 parts:

- General Instructions (*I want you to classify text for me. Your response must always be a python-like list of the 3 most possible class labels like: [0, 12, 15]
See below all the possible intents and their unique labels. Use the description to infer information about each class.*)  
- All possible Intents and their label (unique id) and a **basic description** (eg *This intent is about obtaining flight information*)
- Text to be classified with a prompt to provide the top 3 target labels as a python list    

I am using **jinja2** yet again to define the backbone of my prompt and pass classes (labels) dynamically. 

Prompt template for few-shot classifier: [zero-shot-template.jinja2](../data/zero_shot_template.jinja2)
