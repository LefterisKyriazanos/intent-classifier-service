## Load - GPT Classifier  

### The load function is responsible for iniatializing the model before running the server.  

In more detail the model loader (load function) is responsible for:  

1. preprocessesing of training and test datasets  
2. choosing training data (examples) in case `few-shot` classifier is selected  (see [data_management]              (./data_management.ipynb)).  
3. creating json like representations of classes (labeled intents)  
4. choosing the most representative test_ds based on `test_size` (usually 40-50 records)  (see [data_management]              (./data_management.ipynb)).    
5. creating evaluation metrics based on model performance on test data  (see [evaluation_metrics]              (./evaluation_metrics.ipynb))     
6. deciding status ('SUCCESS', 'FAIL') based on evaluation thresholds.  


more at [gpt_intent_classifier](../gpt_intent_classifier.py) -  `load` method
